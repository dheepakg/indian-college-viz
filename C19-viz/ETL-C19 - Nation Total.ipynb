## Read [JSON file](https://api.covid19india.org/v4/min/timeseries.min.json) from github

```python
df_TT = df['TT'].to_frame()
df_TT_1 = json_normalize(df['TT'], max_level=0)
df_TT_1
```

```python
df_TT_1['2021-05-15'][0]['delta']
```

```python
# Taking transpose to make rows > columns

df_T = df_TT.transpose()
df_T
```

```python
df_T_1 = json_normalize(data = df_T['dates'], max_level=0)
df_T_2 = df_T_1.T
df_T_2.head(5)
```

```python
df_T_2_1 = df_T_2.reset_index()
df_date = df_T_2_1['index'].reset_index()
df_date.columns = ['index', 'date'] 
df_date.head(5)
```

```python
df_measures_1 = json_normalize(data = df_T_2_1[0],max_level=1)
df_measures = df_measures_1.reset_index()
df_measures
```

## Joining the Dataframes

**df_data** consists of measures  
**df_date** consists of dimension   
We need to join these 2 DFs

```python
df_final_0 = pd.merge(
    left = df_measures ,
    right = df_date,
    how="inner",
    on='index',
    suffixes=("_x", "_y")
)
df_final_0.head(5)
```

## Defaulting all numeric columns to 0 (removing NaN)

```python
df_final = df_final_0.apply(lambda x:x.fillna(0) if x.dtype.kind in 'biufc' else x.fillna('.'))
df_final.head(5)
```

## Calculating Active Cases

```python
df_final['delta.active'] = df_final['delta.confirmed'].fillna(0) - df_final['delta.recovered'].fillna(0) - df_final['delta.deceased'].fillna(0)
df_final['delta7.active'] = df_final['delta7.confirmed'].fillna(0) - df_final['delta7.recovered'].fillna(0) - df_final['delta7.deceased'].fillna(0)
df_final['total.active'] = df_final['total.confirmed'].fillna(0) - df_final['total.recovered'].fillna(0) - df_final['total.deceased'].fillna(0)

df_final.head(5)
```

## Reordering Columns

```python
df_final.columns.values
```

```python
df_ordered = df_final[['date', 
        'total.active','total.confirmed', 'total.recovered', 'total.deceased', 'total.tested','total.vaccinated',
        'delta.active','delta.confirmed', 'delta.recovered', 'delta.deceased', 'delta.tested','delta.vaccinated',
        'delta7.active','delta7.confirmed', 'delta7.recovered', 'delta7.deceased', 'delta7.tested','delta7.vaccinated',]]
df_ordered.head(5)
```

## Writing into CSV

```python
df_ordered.to_csv('Total_stats-2021-05-16.csv',index=False)
```
